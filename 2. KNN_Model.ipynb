## Model Training

```python
X_train = []
y_train = []

for label in os.listdir("knn_assignment/train"):
    folder_path = os.path.join("knn_assignment/train", label)
    for img_path in glob.glob(folder_path + "/*.jpg"):
        img = Image.open(img_path).convert("L")  # Convert to grayscale
        img = img.resize((32, 32))  # Resize all to 32x32
        X_train.append(np.array(img).flatten())  # Flatten to 1D
        y_train.append(int(label))

X_train = np.array(X_train)
y_train = np.array(y_train)

print("Training samples:", X_train.shape)
```
